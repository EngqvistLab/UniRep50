```python
import numpy as np
import pandas as pd

from Bio import SeqIO

from unirep.run_inference import BatchInference
import unirep

import tensorflow as tf




np.random.seed(14)

tf.__version__
```

### Get embeddings by batching sequences of same length together

```python
tf.keras.backend.clear_session()

inf_obj = BatchInference(batch_size=500)
df_batch = inf_obj.run_inference(filepath='test.fasta')
```

```python
df_batch
```

### Get embeddings one sequence at a time

```python
tf.keras.backend.clear_session()

bab =  unirep.unirep.babbler1900()

data,index = [],[]
for rec in SeqIO.parse('test.fasta','fasta'):
    seq = rec.seq
    if not bab.is_valid_seq(seq): 
        continue

    av, fh, fc = bab.get_rep(seq)
    lst = list(av)+list(fh)+list(fc)
    data.append(lst)
    index.append(rec.id)

data = np.array(data)

columns = ['av_{0}'.format(i+1) for i in range(1900)] + ['fh_{0}'.format(i+1) for i in range(1900)] + ['fc_{0}'.format(i+1) for i in range(1900)]
df_individual = pd.DataFrame(data=data,index=index, columns=columns)
```

```python
df_individual
```

### Compare the two

```python
df_diff = df_batch - df_individual
df_diff
```

```python
df_diff.abs().max().max()
```
